# Import Required Libraries
Import necessary libraries such as pandas, numpy, seaborn, matplotlib, and scikit-learn.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

# Load and Concatenate Datasets
Load the datasets using pandas and concatenate them into a single DataFrame.

```python
# Load and Concatenate Datasets

# Load the datasets using pandas
df_peugot_1 = pd.read_csv("Dataset/peugeot_207_01.csv", delimiter=';')
df_peugot_2 = pd.read_csv("Dataset/peugeot_207_02.csv", delimiter=';')
df_corsa_1 = pd.read_csv("Dataset/opel_corsa_01.csv", delimiter=';')
df_corsa_2 = pd.read_csv("Dataset/opel_corsa_02.csv", delimiter=';')

# Print the shape of each dataset to understand their sizes
print(df_peugot_1.shape)
print(df_peugot_2.shape)
print(df_corsa_1.shape)
print(df_corsa_2.shape)

# Concatenate the datasets into a single DataFrame
df_new = pd.concat([df_peugot_1, df_peugot_2, df_corsa_1, df_corsa_2], axis=0)

# Display the shape of the concatenated DataFrame
print(df_new.shape)

# Display the first few rows of the concatenated DataFrame to verify the concatenation
df_new.head()
```

# Data Cleaning and Preprocessing
Handle missing values, convert data types, and preprocess the data for analysis.

```python
# Data Cleaning and Preprocessing

# Handle missing values by filling them with 0
df_new = df_new.fillna(0)

# Function to process column values and convert them to float
def process_col_values(df, col_name):
    col_list = list(df[col_name])
    processed_value = []
    for each_value in col_list:
        if each_value == 0:
            processed_value.append(0)
        elif type(each_value) == str:
            processed_value.append(float(each_value.replace(",", ".")))
        else:
            processed_value.append(float(each_value))
    df[col_name] = processed_value

# List of columns to change to float
cols_to_change = ["AltitudeVariation", "VehicleSpeedInstantaneous", "VehicleSpeedAverage", "VehicleSpeedVariance", "VehicleSpeedVariation",
                  "LongitudinalAcceleration", "EngineLoad", "EngineRPM", "MassAirFlow", "VerticalAcceleration", "FuelConsumptionAverage"]

# Process each column in the list
for col in cols_to_change:
    process_col_values(df_new, col)

# Verify the changes by displaying the DataFrame info
df_new.info()

# One-hot encode categorical columns if they exist
categorical_columns = ['roadSurface', 'traffic', 'drivingStyle']
existing_categorical_columns = [col for col in categorical_columns if col in df_new.columns]

if existing_categorical_columns:
    one_hot_encoding = pd.get_dummies(df_new[existing_categorical_columns])
    df_processed = df_new.drop(existing_categorical_columns, axis=1)
    df_processed = pd.concat([df_processed, one_hot_encoding], axis=1)
else:
    df_processed = df_new.copy()

# Shuffle the DataFrame to ensure random distribution
df_processed = shuffle(df_processed)

# Split the data into features (X) and target (y)
X = df_processed.drop(existing_categorical_columns, axis=1, errors='ignore')
y = df_new[existing_categorical_columns]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Normalize the feature data using Min-Max scaling
min_max_scaler = preprocessing.MinMaxScaler()
X_train_scaled = min_max_scaler.fit_transform(X_train)
X_test_scaled = min_max_scaler.transform(X_test)

# Display the shapes of the training and testing sets
print("Training set shape:", X_train_scaled.shape, y_train.shape)
print("Testing set shape:", X_test_scaled.shape, y_test.shape)
```

# Data Normalization
Normalize the data using MinMaxScaler from scikit-learn.

```python
min_max_scaler = preprocessing.MinMaxScaler()

# Fit the scaler on the training data and transform both training and testing data
X_train_scaled = min_max_scaler.fit_transform(X_train)
X_test_scaled = min_max_scaler.transform(X_test)

# Display the shapes of the scaled training and testing sets
print("Training set shape:", X_train_scaled.shape, y_train.shape)
print("Testing set shape:", X_test_scaled.shape, y_test.shape)

# Display the first few rows of the scaled training data to verify normalization
print("First few rows of the scaled training data:")
print(pd.DataFrame(X_train_scaled, columns=X_train.columns).head())

# Display the first few rows of the scaled testing data to verify normalization
print("First few rows of the scaled testing data:")
print(pd.DataFrame(X_test_scaled, columns=X_test.columns).head())
```

# Model Training and Evaluation
Train a RandomForestClassifier and evaluate its performance using accuracy_score.

```python
# Model Training and Evaluation

# Initialize the RandomForestClassifier
forest = RandomForestClassifier(random_state=1)

# Initialize the MultiOutputClassifier with the RandomForestClassifier
multi_target_forest = MultiOutputClassifier(forest, n_jobs=-1)

# Train the model on the scaled training data
multi_target_forest.fit(X_train_scaled, y_train)

# Predict the target values for the scaled testing data
y_test_predicted = multi_target_forest.predict(X_test_scaled)

# Evaluate the model's performance using accuracy_score for each target separately
accuracies = []
for idx, target in enumerate(y.columns):
    accuracy = accuracy_score(y_test.iloc[:, idx], y_test_predicted[:, idx])
    accuracies.append(accuracy)
    print(f"Accuracy for {target}: {accuracy:.2f}")

# Calculate the average accuracy across all targets
average_accuracy = np.mean(accuracies)
print(f"Average Model Accuracy: {average_accuracy:.2f}")

# Visualize feature importance
importances = multi_target_forest.estimators_[0].feature_importances_
feat_importances = pd.Series(importances, index=X_train.columns)
plt.figure(figsize=(10, 6))
feat_importances.sort_values().plot(kind='barh')
plt.title('Feature Importance for Classification')
plt.show()

# Plot confusion matrices for each target
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()
for idx, target in enumerate(y.columns):
    cm = confusion_matrix(y_test.iloc[:, idx], y_test_predicted[:, idx])
    sns.heatmap(cm, annot=True, ax=axes[idx], fmt='d')
    axes[idx].set_title(f'Confusion Matrix - {target}')
plt.tight_layout()
plt.show()

# Print classification report for each target
print("Model Performance Summary:")
for idx, target in enumerate(y.columns):
    print(f"\nResults for {target}:")
    print(classification_report(y_test.iloc[:, idx], y_test_predicted[:, idx]))
```

# Hyperparameter Tuning
Perform hyperparameter tuning to improve model accuracy using GridSearchCV.

```python
# Hyperparameter Tuning

# Import GridSearchCV for hyperparameter tuning
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for RandomForestClassifier
param_grid = {
    'estimator__n_estimators': [50, 100, 150],
    'estimator__max_depth': [None, 10, 20, 30],
    'estimator__min_samples_split': [2, 5, 10],
    'estimator__min_samples_leaf': [1, 2, 4]
}

# Initialize the GridSearchCV with the MultiOutputClassifier and parameter grid
grid_search = GridSearchCV(multi_target_forest, param_grid, cv=3, n_jobs=-1, verbose=2)

# Perform the grid search on the scaled training data
grid_search.fit(X_train_scaled, y_train)

# Get the best parameters from the grid search
best_params = grid_search.best_params_
print(f"Best Parameters: {best_params}")

# Train the model with the best parameters
best_forest = RandomForestClassifier(
    n_estimators=best_params['estimator__n_estimators'],
    max_depth=best_params['estimator__max_depth'],
    min_samples_split=best_params['estimator__min_samples_split'],
    min_samples_leaf=best_params['estimator__min_samples_leaf'],
    random_state=1
)
best_multi_target_forest = MultiOutputClassifier(best_forest, n_jobs=-1)
best_multi_target_forest.fit(X_train_scaled, y_train)

# Predict the target values for the scaled testing data using the best model
y_test_predicted_best = best_multi_target_forest.predict(X_test_scaled)

# Evaluate the best model's performance using accuracy_score
best_accuracy = accuracy_score(y_test, y_test_predicted_best)
print(f"Best Model Accuracy: {best_accuracy:.2f}")

# Visualize feature importance for the best model
best_importances = best_multi_target_forest.estimators_[0].feature_importances_
best_feat_importances = pd.Series(best_importances, index=X_train.columns)
plt.figure(figsize=(10, 6))
best_feat_importances.sort_values().plot(kind='barh')
plt.title('Feature Importance for Best Model')
plt.show()

# Plot confusion matrices for each target using the best model
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()
for idx, target in enumerate(y.columns):
    cm = confusion_matrix(y_test.iloc[:, idx], y_test_predicted_best[:, idx])
    sns.heatmap(cm, annot=True, ax=axes[idx], fmt='d')
    axes[idx].set_title(f'Confusion Matrix - {target}')
plt.tight_layout()
plt.show()

# Print classification report for each target using the best model
print("Best Model Performance Summary:")
for idx, target in enumerate(y.columns):
    print(f"\nResults for {target}:")
    print(classification_report(y_test.iloc[:, idx], y_test_predicted_best[:, idx]))
```
